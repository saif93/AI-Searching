```python
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
Task:     implementation of  8Puzzle using BFS & DFS
Date:           Oct 8, 2016 11:07:01 AM
Programmed By:  Muhammad Saif ul islam
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
import sys
from copy import deepcopy

class EightPuzzleSolver:
    
    def __init__(self,input,goal):
        
        if self.isLegalInput(input) and self.isLegalInput(goal):
            self.initialState = input
            self.goalState = goal
        else:
            print('invalid input or goal state')
    
    def isLegalInput(self, input):
        check = True
        for i in range(len(input)):
            for j in range(len(input)):
                if input[i] == input[j] and i!=j:      #checking repeated Values
                    check = False
                    break
            if input[i]<0 or input[i]>8:               #checking value Range
                check = False
                break
        return check
    
    def isGoalState(self,current):
        if current == self.goalState:
            return True
        else:
            return False
        
    def operatorMove(self,direction,current):   #function to move the tiles
        space = current.index(0)
        if direction == 'left':     #swaping space with left
            current[space],current[space-1] = current[space-1],current[space]
        elif direction == 'right':  #swaping space with right
            current[space],current[space+1] = current[space+1],current[space]
        elif direction == 'up':     #swaping space with up
            current[space],current[space-3] = current[space-3],current[space]
        elif direction == 'down':   #swaping space with down
            current[space],current[space+3] = current[space+3],current[space]
        #self.printCurrentState(current)
            
    def printCurrentState(self,current):
        
        for i in range(len(current)):
            if i%3==0:
                print('\n')
            sys.stdout.write(str(current[i])+" ")
        print('\n-----')
            
           
    def operatorBack(self,previousMove,current): #moving back to previous state
        if previousMove == 'left':
            self.operatorMove('right',current)
        if previousMove == 'right':
            self.operatorMove('left',current)
        if previousMove == 'up':
            self.operatorMove('down',current)
        if previousMove == 'down':
            self.operatorMove('up',current)

    def getDisorderedTiles(self,current):     #get no of tiles out of order as compared to Goal
        counter = 0
        for i in range (len(current)):
            if self.goalState[i]!=current[i]:
                counter+=1
        return counter
        
    def breathFirstSearch(self): #BFS Algorithem
        queue = [] 
        visited = []
        temp = []
        pathCost=0
        visited.append(deepcopy(self.initialState))
        current = deepcopy(self.initialState)
  
        while not self.isGoalState(current): #repeat until goal state is not reached
            space=current.index(0)

            if space not in [0,3,6]: #move left
                temp = deepcopy(current)
                self.operatorMove('left',temp)
                if not temp in visited:
                    queue.append((deepcopy(temp),pathCost+1))
                    visited.append(deepcopy(temp))

            if space not in [2,5,8]: #move right
                temp = deepcopy(current)
                self.operatorMove('right',temp)
                if not temp in visited:
                    queue.append((deepcopy(temp),pathCost+1))
                    visited.append(deepcopy(temp))

            if space not in [0,1,2]: #move up
                temp = deepcopy(current)
                self.operatorMove('up',temp)
                if not temp in visited:
                    queue.append((deepcopy(temp),pathCost+1))
                    visited.append(deepcopy(temp)) 

            if space not in [6,7,8]: #move down
                temp = deepcopy(current)
                self.operatorMove('down',temp)
                if not temp in visited:
                    queue.append((deepcopy(temp),pathCost+1))
                    visited.append(deepcopy(temp))
            aux = queue.pop(0)
            current = aux[0]
            pathCost = aux[1]
        print('Path Cost:',pathCost)
        print('No of Node Visited:',len(visited))
        for i in visited:
            self.printCurrentState(i)
            
    def depthFirstSearch(self): #DFS Algorithm
        stack = [] 
        visited = []
        visited.append(deepcopy(self.initialState))
        current = deepcopy(self.initialState)
  
        while not self.isGoalState(current): #repeat until goal state is not reached
            space=current.index(0)

            if space not in [0,3,6]: #move left
                self.operatorMove('left',current)
                stack.append('left')
                if current  in visited:
                    self.operatorBack(stack.pop(),current)
                else:
                    visited.append(deepcopy(current))
                    continue
            if space not in [2,5,8]: #move right
                self.operatorMove('right',current)
                stack.append('right')
                if current  in visited:
                    self.operatorBack(stack.pop(),current)
                else:
                    visited.append(deepcopy(current))
                    continue
            if space not in [0,1,2]: #move up
                self.operatorMove('up',current)
                stack.append('up')
                if current  in visited:
                    self.operatorBack(stack.pop(),current)
                else:
                    visited.append(deepcopy(current))
                    continue
            if space not in [6,7,8]: #move down
                self.operatorMove('down',current)
                stack.append('down')
                if current  in visited:
                    self.operatorBack(stack.pop(),current)
                else:
                    visited.append(deepcopy(current))
                    continue
            self.operatorBack(stack.pop(),current)
            
        print('Path Cost:',len(stack))
        print('No of Node Visited:',len(visited))
        for i in visited:
            self.printCurrentState(i)
               
```
